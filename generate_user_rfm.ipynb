```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
# Set random seed for reproducibility
np.random.seed(42)
```

```python
# Generate Recency data (bimodal distribution)
recency1 = np.random.exponential(scale=10, size=4000)
recency2 = np.random.exponential(scale=30, size=6000)
recency = np.concatenate([recency1, recency2])
recency = np.clip(recency, 0, 90)  # Clip to match observed range
```

```python
# Generate Frequency data (highly skewed distribution)
frequency = np.random.exponential(scale=50, size=10000)
frequency = np.clip(frequency, 1, 25000)  # Clip to observed range
```

```python
# Generate MonetaryValue data (highly skewed distribution)
monetary_value = np.random.exponential(scale=50000, size=10000)
monetary_value = np.clip(monetary_value, 1, 8e8)  # Clip to observed range
```

```python
# Create DataFrame
df = pd.DataFrame({
    'Recency': recency.astype(int),
    'Frequency': frequency.astype(int),
    'MonetaryValue': monetary_value.astype(int)
})
df
```
